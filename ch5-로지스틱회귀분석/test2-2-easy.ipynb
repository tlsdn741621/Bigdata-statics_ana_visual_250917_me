```python
# 규칙 (Rule): 맨 윗줄은 데이터를 나누는 조건입니다.
# 예를 들어,
# petal length (cm) <= 2.45는
# '꽃잎의 길이가 2.45cm 이하인가?'라는 질문입니다.
# 이 조건이 참(True)이면 왼쪽 자식 노드로, 거
# 짓(False)이면 오른쪽 자식 노드로 이동합니다.
#
# gini (불순도): 해당 노드에 얼마나 다양한 클래스의
# 데이터가 섞여 있는지를 나타내는 지표입니다.
# gini = 0.0이면 해당 노드의 모든 데이터가
# 동일한 클래스로 완벽하게 분류되었음을 의미합니다.
#
# samples (샘플 수): 이 규칙에 해당하는 데이터(샘플)의 총 개수입니다.
#
# value (클래스별 샘플 수):
# 각 클래스(품종)별로 데이터가 몇 개씩 있는지를 보여줍니다.
# 예를 들어, value = [35, 34, 35]는 setosa 35개, versicolor 34개, virginica 35개가 있다는 뜻입니다.
#
# class (예측 클래스):
# 해당 노드에서 가장 많은 비율을 차지하는 클래스입니다.
# 만약 어떤 데이터가 이 노드에서 멈춘다면, 모델은 이 클래스로 예측합니다.
#
# 이처럼 결정 트리를 시각화하면,
# 모델이 붓꽃의 품종을 분류하기 위해
# **'꽃잎의 길이(petal length)'**를 가장 중요한 특징으로 사용하고,
# 그 다음으로 '꽃잎의 너비(petal width)' 등을 기준으로
# 삼는다는 의사결정 과정을 명확하게 파악
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

# 1. 데이터 로드
iris = load_iris()
X = iris.data
y = iris.target

# 2. 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. 결정 트리 모델 생성 및 학습
# max_depth=3: 트리의 최대 깊이를 3으로 제한하여 과적합 방지
dt_model = DecisionTreeClassifier(max_depth=3, random_state=42)
dt_model.fit(X_train, y_train)

# 4. 모델 성능 평가
accuracy = dt_model.score(X_test, y_test)
print(f"모델 정확도: {accuracy:.2f}")

# 5. 결정 트리 시각화
plt.figure(figsize=(15, 10))
plot_tree(dt_model,
          filled=True, # 노드에 색상 채우기
          feature_names=iris.feature_names, # 특징(feature) 이름
          class_names=iris.target_names, # 클래스(품종) 이름
          rounded=True,
          fontsize=10)
plt.title("Decision Tree for Iris Classification")
plt.show()
```
